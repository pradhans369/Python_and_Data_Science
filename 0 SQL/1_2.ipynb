## **Constraints**

Constraints in databases are rules or condition that must be met for data to be inserted, updated, or deleted in a database table. They are used to enforece the integrity of the data stored in a database and to preccent data from becoming inconststent or corrupted.

- **NOT NULL:** This constraint ensures that a column cannot contain null values.
- **UNIQUE:** This constraint ensures that all values in a column are unique.
- **PRIMARY KEY:** This constraint ensures that a column contains unique values and cannot contain null values. It is used to identify and distinguis unique rows in a table.
- **AUTO INCREMENT:** This constraint automatically assigns a unique value to a column when a new row is inserted into the table.
- **CHECK:** This constraint ensures that the values in a column meet a specific condition.
- **DEFAULT:** This constraint assigns a default value to a column when a new row is inserted into the table.
- **FOREIGN KEY:** This constraint ensures that the values in a column match the values in another column in another table. 


Referential Actions : - 
- **RESTRICT:** This action prevents the deletion of a row in a table if there are other rows in another table that reference it.
- **CASCADE:** This action automatically updates or deletes rows in a table when a row in another table is updated or deleted.
- **SET NULL:** This action sets the values in a column to null when a row in another table is updated or deleted.
- **NO ACTION:** This action does nothing when a row in another table is updated or deleted.
