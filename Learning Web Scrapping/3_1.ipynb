### **Here we created another file to collect all the data in the amazon folder**

```python
from bs4 import BeautifulSoup
import os


d = {
    'TITLE' : [1,2],
    'PRICE' : [3,4],
    'LINK' : [5,6]
}

# accessing all the phone names from the amazon folder
NAMES = []
for i in os.listdir("amazon web scrapping data"):
    with open(f"amazon web scrapping data/{i}", encoding="utf-8") as f:
        html_doc = f.read()

    soup = BeautifulSoup(html_doc, "html.parser")
    t = soup.find('h2')
    title = t.get_text(strip=True)
    NAMES.append(title)
    print(title)
```
