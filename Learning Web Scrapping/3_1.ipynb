### **Price**

```python
PRICE = []

for i in os.listdir("amazon web scrapping data"):
    with open(f"amazon web scrapping data/{i}", encoding='utf-8') as f:
        html_doc = f.read()
    
    soup = BeautifulSoup(html_doc, 'html.parser')
    temp = soup.find('span', 'a-price-whole')

    # Adding a basic safety check so the whole loop doesn't crash on out-of-stock items
    if temp is not None:
        price_text = temp.get_text(strip=True)
        # 3. Append to PRICE, not NAMES
        PRICE.append(price_text) 
    else:
        # If the item has no price tag, put "Out of Stock" or "N/A"
        PRICE.append("N/A")

PRICE
```

### **Links**

```python
LINKS = []

for i in os.listdir("amazon web scrapping data"):
    with open(f"amazon web scrapping data/{i}", encoding='utf-8')as f:
        html_doc = f.read()
    
    soup = BeautifulSoup(html_doc, 'html.parser')
    temp = soup.find('a', class_='a-link-normal s-line-clamp-2 puis-line-clamp-3-for-col-4-and-8 s-link-style a-text-normal')

    if temp is not None:
        full_link = "https://www.amazon.in." + temp['href']
        LINKS.append(full_link)          # here we are extracting a link, hence we have to write 'href'
        # print(temp['href'])
    else:
        LINKS.append('N/A')

LINKS
```

### **Doing all the above data extraction in one single for loop**

Whatever we did above, is just for better understanding of visualizing the data

```python
NAME = []
PRICE = []
LINK = []

for i in os.listdir("amazon web scrapping data"):
    with open(f"amazon web scrapping data/{i}", encoding='utf-8') as f:
        html = f.read()
    
    soup = BeautifulSoup(html, 'html.parser')
    
    # -----------------------------NAME------------------------------------
    temp_name = soup.find('h2')
    if temp_name is not None:
        NAME.append(temp_name.get_text(strip=True))
    else:
        NAME.append('N/A')

    # --------------------------PRICE------------------------------------
    temp_price = soup.find('span', 'a-offscreen')
    # temp_price = temp_price.find('span', 'a-offscreen')
    if temp_price is not None:
        PRICE.append(temp_price.get_text(strip=True))
    else:
        PRICE.append('N/A')

    # --------------------------PRICE------------------------------------
    temp_link = soup.find('a', class_='a-link-normal s-line-clamp-2 puis-line-clamp-3-for-col-4-and-8 s-link-style a-text-normal')
    if temp_link is not None:
        full_link = "https://www.amazon.in." + temp_link['href']
        LINK.append(full_link)
    else:
        LINK.append('N/A')


# making a dictionary
dict = {
    'NAME' : NAME,
    'PRICE' : PRICE,
    'LINK' : LINK
}


import pandas as pd
data = pd.DataFrame(dict)
data.sample(10)
```
