```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

url = "https://www.credaiorissa.com/member.php?city=8"
web = requests.get(url, headers=headers)


print(web)
```

### **Company name**

```python
soup = BeautifulSoup(web.content, "html.parser") 

a = soup.find_all('a', class_="text-gray")
company = [i.get_text(strip=True) for i in a]

print(len(company))
print(company)
```

### **Director**

```python
box = soup.find_all('div', "col-md-10 text-cen")

directors = []
for i in box:
    first_h3 = i.find('h3')
    
    if first_h3:  # Good practice to check if it actually exists before calling .get_text()
        directors.append(first_h3.get_text(strip=True))

print(len(directors))
print(directors)
```

### **Address**

```python
address = []

for i in box:
    add = i.find_all('p')
    address.append(add[2].get_text(strip=True))


print(len(address))
print(address)
```

### **Phone**

```python
a = soup.find_all('div', class_="col-md-5")

phone=[]

for i in a:
    paragraphs = i.find_all('p')
    phone.append(paragraphs[0].get_text(strip=True))

print(len(phone))
print(phone)


# we are the deleting the very last element from the list
del phone[-1] 
print(len(phone))
print(phone)
```

### **E-Mail**

```python
temp = soup.find_all('div', class_="col-md-5")

email = []

for i in temp:
    paragraphs = i.find_all('p')
    email.append(paragraphs[1].get_text(strip=True))

print(len(email))
print(email)

del email[-1]
print(len(email))
print(email)


```

## **Converting all the lists into a Dataframe**

```python
data = {
    # 'SL NO.' : SL_NO,
    'COMPANY' : company,
    'ADDRESS' : address,
    'MANAGING DIRECTOR' : directors,
    'PHONE' : phone,
    'E-MAIL' : email
}


df = pd.DataFrame(data)

# df.to_excel('Companies data.xlsx', index=False)

# Create and open the writer
writer = pd.ExcelWriter('Companies data.xlsx', mode='a', engine='openpyxl')
# a = stands for append


# Put the dataframe into the writer
df.to_excel(writer, sheet_name='Balasore', index=False)

# YOU MUST CLOSE IT TO ACTUALLY SAVE IT TO YOUR COMPUTER!
writer.close()
```
