# Making a function that would give recent purchases of an individual in Amazon

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
def recent_transactions(df, user_col, user_id, transactions, date, items, days=30):

    # ASSUMPTIONS
        # df            -- data frame
        # user_col      -- name of the user_id col
        # user_id       -- a feature in df is already in numeric form
        # transactions  -- df holds a feature namd 'transactions' which contains all the transaction amount by the user
        # date          -- df holds a feature named 'date' which contains all the transactions made by the user on the specific date
        # items         -- df holds a feature named 'items' containing all the item names that the user had purchased

    if user_id == False:
        return

    # date from which transactions will be counted
    cutoff_date = pd.to_datetime('today') - timedelta(days = days)

    
    check = (df[user_col] == user_id) & (df[date] > cutoff_date)

    return df.loc[check, [user_col, transactions, date]]
```
