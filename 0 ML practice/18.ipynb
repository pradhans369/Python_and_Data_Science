### Making a function that would give recent purchases of an individual in Amazon

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
def recent_transactions(df, user_col, user_id, transactions, date, items, days=30):

    # ASSUMPTIONS
        # df            -- data frame
        # user_col      -- name of the user_id col
        # user_id       -- a feature in df is already in numeric form
        # transactions  -- df holds a feature namd 'transactions' which contains all the transaction amount by the user
        # date          -- df holds a feature named 'date' which contains all the transactions made by the user on the specific date
        # items         -- df holds a feature named 'items' containing all the item names that the user had purchased

    if user_id == False:
        return

    # date from which transactions will be counted
    cutoff_date = pd.to_datetime('today') - timedelta(days = days)

    
    check = (df[user_col] == user_id) & (df[date] >= cutoff_date)

    return df.loc[check, [user_col, transactions, date]]
```

### The "VIP Customer" Filter

Scenario: You have a dataframe containing a history of purchases. You want to extract the **email addresses** of users who made a large purchase, but only if that purchase was made using a specific payment method.

Your Task: Write a function `get_vip_emails(df, amount_col, payment_col, email_col, min_amount)` that returns a list of emails for users who:

1. Spent greater than or equal to `min_amount`.
2. Paid using exactly `"Credit Card"`.

> (Hint: You will need to combine two conditions in your mask using the & symbol, just like we did in the last problem!)

```python
def get_vip_email(df, amount_col, payment_col, email_col, min_amount):

    if df.empty:
        return 
    

    mask = ([df[amount_col] >= min_amount]) & (df[df[payment_col] == 'Credit Card'])

    return df.loc[mask, email_col].tolist()
```

### The "Price Penalty" (Modifying Data)

Scenario: A shipping company charges a `$5` penalty fee for any package that weighs more than `20 lbs`. You need to update the dataframe to add $5 to the shipping cost of only the heavy packages.

Your Task: Write a function `add_heavy_penalty(df, weight_col, cost_col)` that permanently modifies the `df`.

- If a row's weight is > 20, add 5 to that row's `cost_col`.
- Do not return anything; just modify the dataframe in place.

>(Hint: You can use `df.loc[mask, column_name] = ...` to overwrite data for specific rows!)

```python
def add_heavy_penalty(df, weight_col, cost_col):

    if df.empty:
        return

    mask = (df[weight_col] > 20)

    df.loc[mask, cost_col]= df.loc[mask, cost_col] + 5
```

### Extracting the "Top 3" using `.iloc`

Scenario: You have a dataframe of students ranked by their test scores, already sorted from highest score to lowest score. You need to pull the names of the students who won 1st, 2nd, and 3rd place, but you don't know their exact index labels.

Your Task: Write a function `get_podium_winners(df, name_col)` that returns a list of the names of the top 3 students.

> (Hint: Since you don't care about the row labels and just want the literal first 3 rows from the top of the spreadsheet, this is a perfect time to use `.iloc` and Python slicing `[0:3]`).

```python
def get_podium_winners(df, name_col):

    return df[name_col].iloc[:3].tolist()
```
