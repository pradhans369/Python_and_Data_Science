### The "Top Earner" (GroupBy + Sorting)
**Scenario:** You are given an `employees` dataframe with columns: `['Emp_ID', 'Name', 'Department', 'Salary']`. The CEO wants to know exactly who makes the most money in each department.
**Your Task:** Write a function `get_top_earners(df)` that returns a dataframe showing only the **Name** and **Salary** of the single highest-paid employee in **each** Department.
> *(Hint: You will need to group by the department, but look into `.idxmax()` or sorting the values first and using `.drop_duplicates(subset=['Department'])`!)*


```python
import pandas as pd
import numpy as np
```

```python
def get_top_earners(df):
    
    df_sorted = df.sort_values(by='Salary', ascending=False)
       
    top_earners = df_sorted.drop_duplicates(subset=['Department'], keep='first')
    
    # Step 3: Return just the Name and Salary columns
    return top_earners[['Name', 'Salary']]
```

```python
def get_top_earners(df):

    departments = df['Department'].unique().tolist()

    name = []
    salary = []
    for dpt in departments:
        data = df[df['Department'] == dpt]

        max_salary = data['Salary'].max()
        salary.append(max_salary)

        name_received = data[data['Salary'] == max_salary]['Name'].iloc[0]
        name.append(name_received)



    temp = pd.DataFrame({'name':name, 'max_salary':max_salary})

    return temp
```

### The "Ghost Customers" (Merging & Filtering)
**Scenario:** You have two dataframes. 
1. `customers` df with columns: `['Customer_ID', 'Name']`
2. `orders` df with columns: `['Order_ID', 'Customer_ID', 'Amount']`
**Your Task:** Write a function `find_ghost_customers(customers_df, orders_df)` that returns a list of the **Names** of customers who have **never placed a single order**.
> *(Hint: Look into `pd.merge()`. If you do a "left" merge, customers without orders will have `NaN` in the Order_ID column. You can then use `.isna()` to filter them out!)*


```python
def find_ghost_customers(customers_df, orders_df):

    merged_df = customers_df.merge(orders_df, on='Customer_ID', how='left')

    condition = (merged_df['Order_ID'].isna())    
    names = merged_df.loc[condition, 'Name']                # since .loc returns a list hence there no need to create another list and appending it to that list

    return names
```
