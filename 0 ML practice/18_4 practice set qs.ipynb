### Intermediate Question 1: The "Messy Text Cleaner" (String Manipulation)
**Scenario:** You scraped a dataframe of `product_reviews` with columns `['User', 'Review_Text']`. The text is messy, filled with mixed capitalization. You need to find all reviews that mention the word "battery", regardless of how it is capitalized.

**Your Task:** Write a function `find_battery_complaints(df)` that returns a dataframe containing only the rows where the `Review_Text` contains the word "battery" (e.g., it should catch "Battery", "BATTERY", or "battery").

> *(Hint: Look into the Pandas `.str` accessor, specifically `.str.lower()` and `.str.contains()`. You can use these to build a True/False mask!)*


```python
import numpy as np
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
def find_battery_complaints(df):
    mask = df['Review_Text'].str.lower().str.contains('battery')

    return df.loc[mask]
```

### Intermediate Question 2: The "Executive Summary" (Pivot Tables)
**Scenario:** You have a long, flat `sales` dataframe with columns `['Date', 'Store_Location', 'Daily_Revenue']`. The CEO doesn't want to read a long vertical list. They want a wide spreadsheet where the rows are the Dates, and each Store has its own dedicated column.

**Your Task:** Write a function `create_sales_matrix(df)` that returns a reshaped dataframe where the index is the `Date`, the columns are the `Store_Location`, and the values inside the cells are the `Daily_Revenue`.

> *(Hint: Pandas has a built-in function specifically for this called `.pivot_table()`. You just need to tell it what you want the index, columns, and values to be!)*


```python
def create_sales_matrix(df):
     
    data = pd.pivot_table(
        df, 
        index='Date',                 # The rows
        columns='Store_Location',     # The columns
        values='Daily_Revenue'        # The numbers inside the table
    )

    return data
```

### Intermediate Question 4: The "Health Categorizer" (Custom Apply Logic)
**Scenario:** You have a `patients` dataframe with columns `['Patient_ID', 'Height_m', 'Weight_kg']`. You need to calculate the BMI for each patient, and then create a new column called `'Status'` that labels them as "Normal" or "Overweight" based on that BMI.

**Your Task:** Write a function `categorize_patients(df)` that calculates the BMI `(Weight / Height^2)`. If the BMI is strictly greater than `25.0`, set their Status to "Overweight". Otherwise, set it to "Normal".

> *(Hint: You can use `import numpy as np` and use `np.where(condition, 'Overweight', 'Normal')`. It works exactly like an IF statement for entire columns at once!)*


```python
def categorize_patients(df):
    weight = df['Weight_kg']
    height = df['Height_m']

    BMI = weight / (height * height)

    condition = (BMI > 25.0)
    
    df['status'] = np.where(condition, 'Overweight', 'Normal')

    return df
```

```python
def categorize_patients(df):
    weight = df['Weight_kg']
    height = df['Height_m']

    BMI = weight / (height * height)

    df['status'] = BMI

    df['status'] = df['status'].apply(lambda x : 'Overweight' if x > 25.0 else 'Normal')

    return df
```
