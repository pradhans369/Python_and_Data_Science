```python
class data:
    def __init__(self, name, id, job, location):
        self.name = name
        self.id = id
        self.job = job
        self.__location = location
    

    def print_data(self):
        # print(f" Name : {self.name}\n ID : {self.id}\n JOB : {self.job}\n Location : {self.location}")
        print(f"PRINTING PRIVATE VARIABLE : {self.__location}")
```

```python
class userdetails:
    def __init__(self, name, email, password, phone, role):
        self.name = name
        self.email = email
        self.__password = password
        self.__phone = phone
        self.role = role
    
    # printing all the details
    def check(self):
        print(f" USER : {self.name}\n E-MAIL : {self.email}\n, PASSWORD : {self.get_data()[0]}\n PHONE : {self.get_data()[1]}")
    
    # function to print the private variables
    def get_data(self):
        return self.__password, self.__phone
    
    # function to change a variable
    def change_name(self, name):
        self.name = name
```

---

```python
a = userdetails('Ashura', 'ashura123@gmail.com', 'ashura12345', 756481548, 'Data Scientist')

a.check()
```

```python
a.change_name('KING')

a.check()
```

### Using `@property`

Cons :
- You can not pass any parameters into it


```python
class maths:
    def __init__(self, sub_a, sub_b, sub_c):
        self.sub_a = sub_a
        self.sub_b = sub_b
        self.sub_c = sub_c
        self.avg = str((self.sub_a + self.sub_b + self.sub_c) / 3) + "%"
    
    @property
    def average(self):
        return self.avg
    
    def average2(self):
        return self.avg
```

```python
a = maths(50,40,30)


print(a.average2())
print(a.sub_a)
print(a.average)        # when you property you can use any function like an attribute. Just like "a.sub_a"
```

```python
print(a.sub_a)
print(a.average)
```

```python
class Complex:
    def __init__(self, A, B):
        self.A = A
        self.B = B
    
    def show(self):
        print(self.A, 'i +', self.B, 'j')

    # type 1
    def add(self, x):
        newA = self.A + x.A 
        newB = self.A + x.B 

        return Complex(newA, newB)

    # type 2
    def add2(self, x, y):
        newA = x.A + y.A
        newB = x.B + y.B 

        return Complex(newA, newB)
```

```python
print(1 + 2)
print('my' + 'school')
print([1,2,3] + [4,5,6]) 
```

```python
n1 = Complex(1, 3)
n1.show()
print()
```

### Using `del`

```python
class userdetails:
    def __init__(self, name, email, password):
        self.name = name
        self.email = email
        self.password = password
        self.__phone = None
        self.__role = None
    
    def set_phone_role(self, phone=None, role=None):
        self.__phone = phone
        self.__role = role

        return self.__phone, self.__role
    
    def show(self):
        print(f"Name: {self.name}")
        print(f"Email: {self.email}")
        print(f"Password: {self.password}")
        
        if self.__phone is not None or self.__role is not None:
            print(f"Phone: {self.__phone}")
            print(f"Role: {self.__role}")
```

```python
u1 = userdetails('Ashura', 'ashura45@gmail.com', 'ashura231512')

print(u1.show())
```

```python
u1.set_phone_role(role='engineer')

print(u1.show())
```

```python
u1.set_phone_role(phone=43166255165, role='scientist')

print(u1.show())
print()
u1.show()
```
