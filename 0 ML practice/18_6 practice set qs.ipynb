### The "Machine Learning Prep" (.map)
**Scenario:** You have a `loan_applications` dataframe with a column called `['Status']`. The column contains three words: `"Approved"`, `"Rejected"`, and `"Pending"`. You need to prepare this data for a Machine Learning model, which can only read 1s and 0s. 

**Your Task:** Write a function `prep_loan_data(df)` that translates `"Approved"` into the number `1`, translates `"Rejected"` into the number `0`, and destroys the `"Pending"` applications by turning them into `NaN`. 

> *(Hint: Since you want to perfectly translate exact words into numbers, and you WANT to destroy the unmatched "Pending" rows, `.map()` is the absolute perfect tool for this!)*


```python
def prep_loan_data(df):
    replace({
        'Approved':1,
        'Rejected':0       # if you write 'Pending':'NaN', pandas will assign the text NaN to Pending class, which will be in a text format
                           # if you do not mention a class, then map will automatically assign 'NaN' to all the not mentioned class
    })

    df['Status'] = df['Status'].map(replace)

    return df
```

### The "Dynamic Pricing" (np.where)
**Scenario:** You work for an airline. You have a `flights` dataframe with two columns: `['Days_Until_Flight', 'Ticket_Price']`. The airline wants to implement "surge pricing" for last-minute buyers.

**Your Task:** Write a function `apply_surge_pricing(df)` that permanently updates the `Ticket_Price` column. If the `Days_Until_Flight` is strictly less than `7`, the new ticket price should be the old price multiplied by `1.20`. If it is 7 days or more, the price should stay exactly the same.

> *(Hint: Because you need to apply mathematical greater-than/less-than logic across two different columns, you must use `np.where()`. Think about what goes in the "What to do if True" slot vs the "What to do if False" slot!)*


```python
def apply_surge_pricing(df):
    
    condition = (df['Days_Until_Flight'] < 7)
    price_hike = 1.20
    
    df['Ticket_Price'] = np.where(condition, df['Ticket_Price'] * price_hike, df['Ticket_Price'])

    return df
```

```python
def apply_surge_pricing(df):
    
    condition = (df['Days_Until_Flight'] < 7)
    price_hike = 1.20
    
    df.loc[condition, 'Ticket_Price'] = df.loc[condition, 'Ticket_Price'] * price_hike

    return df
```

### The "Database Crusher" (.melt)
**Scenario:** Your accounting department handed you a terribly formatted `store_sales` dataframe. It is a wide table with 4 columns: `['Store_Name', 'Jan_Sales', 'Feb_Sales', 'Mar_Sales']`. A Data Scientist cannot run a model on a wide table.

**Your Task:** Write a function `unpivot_sales(df)` that crushes this wide table into a long, vertical list with exactly 3 columns: `['Store_Name', 'Month', 'Total_Sales']`. 

> *(Hint: You will need to use `pd.melt()`. The `id_vars` are the columns you want to lock in place (Store_Name), and the `value_vars` are the columns you want to crush into a single vertical list!)*


```python
def unipivot_sales(df):
    df = df.melt(id_vars='Store_Name', value_vars=['Jan_Sales', 'Feb_Sales', 'Mar_Sales'], var_name='Month', value_name='Total_sales')

    return df
```
