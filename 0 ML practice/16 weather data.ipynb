# **Weather**

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('london weather AQI.csv')
df.head(5)
```

```python

# checking for null values
print(df.isna().sum())
```

```python
print(df.columns)
cols = df.columns
```

```python
df.info()
```

```python
# converting dt_txt into date time format

df['dt_txt'] = pd.to_datetime(df['dt_txt'])
print(df['dt_txt'].info())

df['dt_txt']
```

```python
df['date'] = df['dt_txt'].dt.date
df['time'] = df['dt_txt'].dt.time

# dropping useless features

df = df.drop(columns=['Unnamed: 0', 'dt', 'dt_txt'])


df.head(5)
```

```python
df.describe()
```

```python
# df.columns

df['summery'].to_list()

```

```python
# doing feature extraction

# first we have to convert the 'summery featue' into a string
df['summery'] = df['summery'].astype(str)
# df['summery'].str.extract(r'^([^,]+)', expand=False)          # doing the same with regex

df['weather type'] = df['summery'].str.split(',').str[0]

# dropping the summery column
df = df.drop(columns=['summery'])

print(df['weather type'].unique())

df['weather type']
```

```python
# since date and time are still in object format we have to extract day, month, and year
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()  
# df['month'] = df['date'].dt.month         # this will put the months in numbers, use this while training or outlier detection
df['date'] = df['date'].dt.day

df.head()
```
