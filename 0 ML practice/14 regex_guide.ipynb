# The Ultimate Guide to Regular Expressions (Regex)

**Regex** (short for Regular Expressions) is a powerful tool used for matching patterns in text. Think of it as "Ctrl+F" on steroids. Instead of searching for a literal word like "apple", you can search for "any word starting with 'a' and ending with 'e'".

## 1. The Basics: How it thinks
Regex reads text from left to right, character by character, trying to match your pattern.

*   **Literal Characters**: `abc` matches the exact text "abc".
*   **The Wildcard**: `.` matches **any single character** (except a new line).
    *   Pattern `h.t` matches "hot", "hat", "hit", "h@t".

## 2. Character Sets `[]`
Use brackets to define a list of allowed characters for a *single* position.

*   `[aeiou]` - Matches any **one** vowel.
*   `[0-9]` - Matches any digit from 0 to 9.
*   `[a-z]` - Matches any lowercase letter.
*   `[^0-9]` - The `^` inside brackets means **NOT**. Matches anything strictly *not* a number.

## 3. Quantifiers (How many?)
These symbols tell Regex how many times the *previous* character/group should appear.

| Symbol | Meaning | Mnemonics (Easy way to remember) |
| :--- | :--- | :--- |
| `*` | **0 or more** | "Maybe it's there, maybe lots of them." |
| `+` | **1 or more** | "At least one must be there." |
| `?` | **0 or 1** | "Optional. Once or nothing." |
| `{3}` | **Exactly 3** | "Exactly 3 times." |
| `{2,4}`| **2 to 4** | "Between 2 and 4 times." |

**Examples:**
*   `colou?r` matches "color" (0 'u's) and "colour" (1 'u').
*   `go+d` matches "god", "good", "goood" (but not "gd").

## 4. Anchors (Where?)
Anchors don't match characters; they match **positions**.

*   `^` - Start of the line.
    *   `^Hello` matches "Hello world" but NOT "Oh, Hello".
*   `$` - End of the line.
    *   `bye$` matches "Good bye" but NOT "bye now".

## 5. Groups `()`
Parentheses group parts of the pattern together so you can modify them or extract specific pieces.

*   `(ha)+` matches "haha", "hahaha". (Without brackets, `ha+` would match "haaa").
*   **OR Operator**: `|` acts like "or".
    *   `gr(a|e)y` matches "gray" or "grey".

## 6. Predefined Shortcuts (The Cheat Sheet)
Regex has shorthand codes for common types.

*   `\d` - Any Digit (0-9)
*   `\w` - Any Word Character (a-z, A-Z, 0-9, and underscore `_`)
*   `\s` - Any Whitespace (space, tab, newline)
*   `\D`, `\W`, `\S` - The Capital version is the **OPPOSITE** (Not digit, Not word, Not space).

## 7. Escaping `\`
If you want to search for a literal dot `.` or question mark `?`, you must put a backslash before it.
*   Pattern `3\.14` matches "3.14". (Pattern `3.14` would act as a wildcard and match "3x14").

## Python Example: `re` module

Python uses the built-in `re` library.

**Key Functions:**
1.  `re.search(pattern, text)`: Finds the *first* match anywhere.
2.  `re.match(pattern, text)`: Checks for a match *only at the beginning*.
3.  `re.findall(pattern, text)`: Returns a **list** of all matches.
4.  `re.sub(pattern, replacement, text)`: Replaces matches with new text.

### Real Breakdown: Validating an Email
Pattern: `^[\w\.-]+@[\w\.-]+\.\w+$`

1.  `^` : Start of line.
2.  `[\w\.-]+` : One or more word characters, dots, or dashes (the username).
3.  `@` : Liberal '@' symbol.
4.  `[\w\.-]+` : Domain name (like gmail, yahoo).
5.  `\.` : Literal dot.
6.  `\w+` : The extension (com, org, net).
7.  `$` : End of line.
