```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target


print(df['target'].unique())
df.head()
```

```python
km = KMeans(n_clusters=3, random_state=42)
df['cluster_id'] = pd.DataFrame(km.fit_predict(df))
# df['cluster_id'] = pd.DataFrame(km.fit_predict(df[['target']]))
# K-Means is an unsupervised algorithm used to find patterns in data when you don't have labels. By passing the target (the answers) to the algorithm, you are basically asking it to cluster the labels themselves. This is trivial and doesn't teach the model anything about the flowers.
df
```

```python
plt.figure(figsize=(18, 4))
plt.subplot(121)
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['cluster_id'], palette=['r', 'g', 'b'])

plt.subplot(122)
sns.scatterplot(x=df['sepal width (cm)'], y=df['petal width (cm)'], hue=df['cluster_id'], palette=['r', 'g', 'b'])
plt.show()
```
