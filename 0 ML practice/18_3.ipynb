```python

def check(df):

    continent = df['continent'].unique().tolist()       # by default unique returns array hence converted into a list

    names = []
    max_longitude = []
    max_latitude = []

    for i in continent:

        # append continent names to 'names'
        if i not in names:
            names.append(i)

        temp = df[df['continent'] == i]

        max_longitude.append(temp['Longitude'].max())
        max_latitude.append(temp['Latitude'].max())

    

    a = pd.DataFrame({
        'names' : names,
        'max_longitude' : max_longitude,
        'max_latitude' : max_latitude
    })

    return a
```

```python
check(df)
```
