# **K Means Clustering**

K-Means Clustering is an **unsupervised machine learning** algorithm used to group unlabeled data into $K$ distinct, non-overlapping clusters. It works by minimizing the variance within each cluster (intra-cluster sum of squares).

### How the Algorithm Works
1.  **Initialization**: Choose the number of clusters $K$ and randomly select $K$ data points as initial centroids.
2.  **Assignment**: Assign each data point to the nearest centroid based on a distance metric (usually Euclidean distance).
3.  **Update**: Calculate the mean of all points assigned to each cluster and move the centroid to this new mean position.
4.  **Repeat**: Iteratively perform steps 2 and 3 until the centroids no longer move significantly or a maximum number of iterations is reached.

### Choosing the Optimal $K$ (The Elbow Method)
To find the best value for $K$, we plot the **Within-Cluster Sum of Squares (WCSS)** against different values of $K$.
- **WCSS**: The sum of squared distances between each point and its assigned centroid.
- **The "Elbow"**: The point where the rate of decrease in WCSS slows down significantly is considered the optimal $K$.

### Mathematical Objective
The goal is to minimize the objective function:
$$J = \sum_{j=1}^{K} \sum_{i=1}^{n} ||x_i^{(j)} - c_j||^2$$
Where:
- $n$ is the number of data points.
- $K$ is the number of clusters.
- $x_i^{(j)}$ is a data point in cluster $j$.
- $c_j$ is the centroid of cluster $j$.

### Advantages
- Simple to understand and implement.
- Efficient on large datasets (linear time complexity).
- Guaranteed convergence.

### Limitations
- Requires pre-specifying the number of clusters $K$.
- Sensitive to initial centroid placement (can get stuck in local optima).
- Sensitive to outliers.
- Struggles with non-spherical shapes or clusters of varying sizes.
