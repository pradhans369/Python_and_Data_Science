# **Decision Tree Visualization with dtreeviz**

This notebook demonstrates how to visualize Decision Trees using the powerful `dtreeviz` library (v2.0+) alongside Scikit-Learn.

```python
import matplotlib.pyplot as plt
import dtreeviz
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from IPython.display import display
```

### **1. Prepare Data and Fit Model**

```python
iris = load_iris()
X = iris.data
y = iris.target

# Fit a Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=3) 
clf.fit(X, y)
```

### **2. Visualize using Scikit-Learn (Built-in)**

```python
plt.figure(figsize=(12, 8))
from sklearn import tree
tree.plot_tree(clf, 
               feature_names=iris.feature_names, 
               class_names=list(iris.target_names), 
               filled=True, 
               rounded=True)
plt.show()
```
