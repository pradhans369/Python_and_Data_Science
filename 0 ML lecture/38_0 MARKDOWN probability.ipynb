# **Probability**

# Complete Guide to Probability: From Zero to Hero

Welcome to the ultimate, easy-to-understand guide on **Probability**. Whether you're a beginner, a student, or a future Data Scientist, this guide will break down complex mathematical concepts into simple, bite-sized pieces using everyday examples.



## 1. What is Probability?

At its simplest, **probability is the measure of how likely an event is to happen.** 

We deal with probability every day:
- *"What's the chance it will rain today?"*
- *"What are the odds my favorite team wins the match?"*
- *"If I buy a lottery ticket, will I win?"*

Probability is measured on a scale from **0 to 1** (or 0% to 100%):
- **0 (or 0%)**: Impossible to happen (e.g., rolling a 7 on a standard 6-sided die).
- **1 (or 100%)**: Certain to happen (e.g., the sun rising tomorrow).
- **0.5 (or 50%)**: Even chance (e.g., flipping a coin and getting Heads).



## 2. The Core Vocabulary (The "Jargon" Made Easy)

Before we do any math, let's understand the words mathematicians use:

1. **Experiment**: An action where the result is uncertain. 
   - *Example: Tossing a coin, rolling a die, drawing a card.*
2. **Outcome**: A single possible result of an experiment.
   - *Example: Getting a "Tails" when you toss a coin.*
3. **Sample Space ($S$)**: The complete list of **ALL** possible outcomes.
   - *Example (Coin Tossing):* $S = \{\text{Heads}, \text{Tails}\}$
   - *Example (Rolling a Die):* $S = \{1, 2, 3, 4, 5, 6\}$
4. **Event ($E$)**: One or more outcomes that you are specifically looking for. 
   - *Example: Rolling an EVEN number on a die. The event covers outcomes $\{2, 4, 6\}$.*



## 3. The Golden Formula of Probability

For simple scenarios (where all outcomes are equally likely), here is the only formula you need:

$$P(\text{Event}) = \frac{\text{Number of Favorable Outcomes}}{\text{Total Number of Possible Outcomes}}$$

### Example 1: Rolling a Die
What is the probability of rolling a 4 on a standard six-sided die?
- Favorable outcome (rolling a 4): exactly 1 outcome.
- Total possible outcomes (1, 2, 3, 4, 5, 6): exactly 6 outcomes.
- **$P(4) = \frac{1}{6}$**

### Example 2: Rolling an Even Number
What is the probability of rolling an even number (2, 4, or 6)?
- Favorable outcomes (2, 4, 6): exactly 3 outcomes.
- Total possible outcomes: 6 outcomes.
- **$P(\text{Even}) = \frac{3}{6} = \frac{1}{2} = 50\%$**



## 4. The 3 Golden Rules (Axioms) of Probability

1. **The Range Rule**: The probability of an event $A$ is always between 0 and 1.
   $$0 \le P(A) \le 1$$
2. **The Sum Rule**: The sum of the probabilities of ALL possible outcomes in the Sample Space exactly equals 1.
   - If you flip a coin: $P(\text{Heads}) + P(\text{Tails}) = 0.5 + 0.5 = 1$.
3. **The Complement Rule**: The probability of an event NOT happening is 1 minus the probability of it happening. We call "NOT A" the complement, written as $A'$ or $A^c$.
   $$P(A') = 1 - P(A)$$
   - *Example:* If the chance of rain is 30% ($0.3$), the chance of it NOT raining is $1 - 0.3 = 0.7$ (70%).



## 5. Types of Events

To solve harder problems, we need to know how events relate to each other.

### A. Independent Events
Two events are independent if the outcome of one **does not affect** the other.
- *Example:* Flipping a coin and rolling a die. Getting "Heads" has zero impact on rolling a "4".

### B. Dependent Events
Two events are dependent if the outcome of the first **DOES affect** the outcome of the second.
- *Example:* Drawing cards from a deck **without replacing them**. 
  - Probability of drawing an Ace first: $4/52$.
  - If you don't put it back, you only have 51 cards left. The probability of drawing a second Ace is now $3/51$.

### C. Mutually Exclusive Events
Events that **cannot happen at the same time**.
- *Example:* Turning left and turning right at the same time. 
- *Example:* Rolling a single die and getting a 2 AND a 5 simultaneously. Impossible!



## 6. Combining Probabilities (The "AND" & "OR" Rules)

### The Addition Rule (The "OR" Rule)
When you want to find the probability of Event A **OR** Event B happening, you add them.

1. **If Mutually Exclusive** (They can't happen together):
   $$P(A \text{ or } B) = P(A) + P(B)$$
   - *Example:* Probability of rolling a 2 OR a 3 on a die $= \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}$.

2. **If NOT Mutually Exclusive** (They CAN happen together, so we must subtract the overlap so we don't count it twice):
   $$P(A \text{ or } B) = P(A) + P(B) - P(A \text{ and } B)$$
   - *Example:* Probability of drawing a King OR a Heart from a deck of cards.
     - $P(\text{King}) = \frac{4}{52}$
     - $P(\text{Heart}) = \frac{13}{52}$
     - $P(\text{King AND Heart}) = \frac{1}{52}$ (The King of Hearts was counted in both!)
     - Total $= \frac{4}{52} + \frac{13}{52} - \frac{1}{52} = \frac{16}{52}$.

### The Multiplication Rule (The "AND" Rule)
When you want to find the probability of Event A **AND** Event B happening one after another, you multiply them.

1. **If Independent**:
   $$P(A \text{ and } B) = P(A) \times P(B)$$
   - *Example:* Flipping two Heads in a row $= \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}$ (or 25%).

2. **If Dependent**:
   $$P(A \text{ and } B) = P(A) \times P(B|A)$$
   *(Note: $P(B|A)$ means "Probability of B happening GIVEN THAT A already happened".)*



## 7. Conditional Probability

**Conditional Probability** is the probability of an event happening **given that** another event has already occurred. It changes our "Sample Space" pool.

Formula:
$$P(A|B) = \frac{P(A \text{ and } B)}{P(B)}$$
*(Read as: The probability of A given that B has occurred).*

**Easy Example:**
In a class of 100 students:
- 40 play Soccer.
- 20 play Soccer AND Basketball.
If I pick a student who plays Soccer (this is the given condition, so our total pool is now just those 40 kids), what is the probability they also play Basketball?
- $P(\text{Basketball } | \text{ Soccer}) = \frac{20 \text{ (Both)}}{40 \text{ (Soccer only)}} = \frac{1}{2} = 50\%$.



## 8. Bayes' Theorem (The Crown Jewel of Data Science)

Named after Reverend Thomas Bayes, this theorem allows us to **update our beliefs/probabilities based on new evidence**. It is the foundation of many Machine Learning algorithms (like the Naive Bayes Classifier).

$$P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}$$

**What do these parts mean?**
- **$P(A)$**: **Prior Probability** (What we believed overally before seeing evidence).
- **$P(B)$**: **Evidence** (The total probability of the new data).
- **$P(B|A)$**: **Likelihood** (The probability of seeing this evidence if A is true).
- **$P(A|B)$**: **Posterior Probability** (Our NEW, updated belief after seeing the evidence).

### The Classic Medical Test Intuition:
A rare disease affects 1% of the population.
A test for the disease is 99% accurate:
- If you have the disease, the test is positive 99% of the time.
- If you don't have it, there's a 2% false positive rate.

If someone gets a positive test, what is the actual chance they have the disease?
Using Bayes' Theorem, it turns out the chance is actually only **around 33%**, not 99%! This is because the disease is so rare to begin with ($P(A)$ prior) that the false positives outnumber the true positives in the whole population. The math protects us from trusting the test blindly!



## 9. Random Variables & Distributions (A Quick Primer)

In Data Science, we don't just calculate probabilities of single events; we look at the probability of ALL possible events mapped out over a graph. This is called a **Probability Distribution**.

### A. Random Variables
A Random Variable assigns a numerical value to each possible outcome.
- **Discrete**: Can only take specific, distinct values (e.g., number of heads in 5 coin tosses: 0, 1, 2, 3, 4, 5). You can "count" them.
- **Continuous**: Can take any value within a range (e.g., the exact height of a person: 170.123 cm). You "measure" them.

### B. Common Probability Distributions
1. **Normal Distribution (The Bell Curve)**: The most famous! Many natural things (height, IQ, test scores) follow a symmetrical bell shape where most data clusters around the average (mean).
2. **Binomial Distribution**: Used for experiments with exactly two outcomes (Success/Failure). *E.g., The probability of getting exactly 3 heads in 10 flips.*
3. **Poisson Distribution**: Measures the probability of a given number of events happening in a fixed time/space. *E.g., How many emails you receive in an hour.*



## 10. Summary Cheat Sheet

| Concept | Easy Definition | Formula / Representation |
| :--- | :--- | :--- |
| **Probability base formula** | Expected items divided by all items | $P(A) = \frac{\text{Favorable}}{\text{Total}}$ |
| **Complement Rule** | The chance of an event NOT happening | $P(A') = 1 - P(A)$ |
| **Addition ("OR") Rule** | Add probabilities (subtract overlaps if any) | $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ |
| **Multiplication ("AND") Rule** | Multiply probabilities together | $P(A \cap B) = P(A) \times P(B\|A)$ |
| **Conditional Probability** | Probability of A *knowing* B already happened | $P(A\|B) = \frac{P(A \cap B)}{P(B)}$ |
| **Bayes' Theorem** | Updating probability rationally with new evidence | $P(A\|B) = \frac{P(B\|A) \times P(A)}{P(B)}$ |

# $$Understanding Dice Probability$$


When working with dice, there are different ways to calculate probabilities depending on exactly what you are trying to find.

### A. Rolling Two Dice: Total Possible Outcomes
If you roll two standard 6-sided dice together, there are exactly **36 total possible outcomes**. 

**Why?**
Using the **Fundamental Counting Principle**:
- Die 1 has 6 possible outcomes (1, 2, 3, 4, 5, 6).
- Die 2 has 6 possible outcomes (1, 2, 3, 4, 5, 6).
- $6 \times 6 = 36$ total combinations.

### B. Probability of Getting an "Even Number" on BOTH Dice (Single Roll)
If you roll two dice once, what is the chance that both dice show an even number (2, 4, or 6)? Let's look at two ways to solve this:

**Method 1: The Multiplication Rule**
- The chance of an even number on Die 1 is 3 out of 6 ($\frac{1}{2}$).
- The chance of an even number on Die 2 is 3 out of 6 ($\frac{1}{2}$).
- Multiply them: $\frac{1}{2} \times \frac{1}{2} = \boldsymbol{\frac{1}{4}}$ **(or 25%)**.

**Method 2: Counting Outcomes (Fundamental Counting Principle)**
- How many even outcomes are possible on Die 1? Just 3 (2, 4, 6).
- How many even outcomes are possible on Die 2? Just 3 (2, 4, 6).
- Outcomes with even on both $= 3 \times 3 = \boldsymbol{9 \text{ winning combinations}}$.
- Probability $= 9 \text{ favorable outcomes} \div 36 \text{ total outcomes} = \boldsymbol{\frac{9}{36} = \frac{1}{4}}$.

### C. Rolling Two Dice, TWICE in Order
What if you roll 2 dice, look at them, pick them up, and roll them a second time? "Getting even numbers" can mean two different things here:

**Scenario 1: You want EVERY single die you roll to be even**
You are looking at 4 different die faces in total. 
Since the chance of any single die landing on even is $\frac{1}{2}$, you use the Multiplication Rule for all 4 events:
$$P(\text{Even on all 4 dice}) = \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} = \boldsymbol{\frac{1}{16}} \text{ (or 6.25\%)}$$

**Scenario 2: You want the SUM of the dice to be even on both rolls**
When adding dice together, an even sum happens in two ways:
- Even + Even = Even ($\frac{1}{4}$ probability)
- Odd + Odd = Even ($\frac{1}{4}$ probability)

So, the total chance of an even sum on a single throw of two dice is $\frac{1}{4} + \frac{1}{4} = \boldsymbol{\frac{1}{2}}$ (50%).

Because you want an even sum on the *first roll* **AND** the *second roll*:
$$P(\text{Even Sum Both Times}) = \frac{1}{2} \times \frac{1}{2} = \boldsymbol{\frac{1}{4}} \text{ (or 25\%)}$$
