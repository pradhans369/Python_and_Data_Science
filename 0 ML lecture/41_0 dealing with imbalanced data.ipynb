# **Imbalanced Data**

### What is Imbalanced Data?

**Imbalanced data** refers to a situation in a classification problem where the distribution of classes is not uniform. In such datasets, one or more classes (the **majority classes**) significantly outnumber the other classes (the **minority classes**).

#### Why is it a problem?
Most machine learning algorithms are designed to maximize overall accuracy. When the data is heavily imbalanced:
*   **Bias toward Majority:** The model may learn to predict the majority class for all instances to achieve high accuracy, ignoring the minority class entirely.
*   **Misleading Metrics:** Accuracy becomes a poor metric. For example, in a dataset where 99% of transactions are legitimate and 1% are fraudulent, a model that predicts "legitimate" for every case will be 99% accurate but useless for fraud detection.

#### Common Examples:
*   **Fraud Detection:** Legitimate transactions far outnumber fraudulent ones.
*   **Medical Diagnosis:** Healthy patients are usually much more common than those with a specific rare disease.
*   **Spam Filtering:** The volume of "ham" (non-spam) emails often outweighs "spam."

#### How to Handle Imbalanced Data:
1.  **Resampling Techniques:** 
    *   *Oversampling:* Increasing the number of minority class samples (e.g., SMOTE).
    *   *Undersampling:* Reducing the number of majority class samples.
2.  **Evaluation Metrics:** Using Precision, Recall, F1-Score, or ROC-AUC instead of Accuracy.
3.  **Algorithmic Approaches:** Using cost-sensitive learning or tree-based algorithms that handle imbalance better.


There are different ways to handle Imbalanced data
- Undersampling
- Oversampling
- SMOTE
- Ensemble Methods
- Cost Sensitive Learning


### 1. Undersampling
Undersampling involves reducing the number of instances in the majority class to match the size of the minority class. This technique is typically used when the dataset is sufficiently large, allowing for a balanced subset to be created without losing the core characteristics of the majority class. However, the primary drawback is the potential loss of valuable information that could have been used to define the majority class's boundaries.

### 2. Oversampling
Oversampling increases the number of instances in the minority class by duplicating existing samples. This ensures that the machine learning algorithm receives enough signals from the minority class to learn its patterns. While it prevents information loss, it significantly increases the risk of overfitting, as the model may simply memorize the duplicated examples rather than learning generalizable features.

### 3. SMOTE (Synthetic Minority Over-sampling Technique)
SMOTE is a more advanced oversampling technique that creates synthetic examples rather than just duplicating them. It works by selecting a point from the minority class, finding its k-nearest neighbors, and generating new points at random intervals along the lines connecting these neighbors. This approach helps the model generalize better by expanding the decision boundary of the minority class in a more continuous manner.

### 4. Ensemble Methods
Ensemble methods handle imbalance by combining the predictions of multiple base classifiers. Techniques like Balanced Random Forest or EasyEnsemble use bagging and boosting variations specifically designed for imbalanced data. These methods often incorporate internal undersampling or oversampling within each bootstrap sample, ensuring that every individual model in the ensemble is trained on a relatively balanced representation of the data.

### 5. Cost Sensitive Learning
Cost Sensitive Learning shifts the focus from balancing the data to modifying the algorithm's objective. It assigns a higher penalty (cost) to misclassifying a minority class instance than a majority class instance. By adjusting the loss function, the model is forced to prioritize the correct classification of the minority class, making it a highly effective strategy when certain types of errors are more expensive or critical than others.



| Technique | Pros | Cons |
| :--- | :--- | :--- |
| **Undersampling** | Reduces training time and computational cost; effective for extremely large datasets. | Risk of losing critical information from the majority class; can lead to biased models. |
| **Oversampling** | No information loss; simple to implement and ensures the minority class is well-represented. | High risk of overfitting due to exact duplicates; increases training time and memory usage, increase in memory |
| **SMOTE** | Reduces overfitting compared to random oversampling; creates more diverse and generalizable boundaries. | Can introduce noise if synthetic points are created between overlapping classes; computationally expensive, doesnot handle categorical data well, dependency of choice of neighbors, sensitive to outliers, balance achieved may not reflect true nature, increase in data size. |
| **Ensemble Methods** | Highly robust and often yields the best predictive performance; handles imbalance internally. | Computationally intensive; complex to tune and harder to interpret than single models. |
| **Cost Sensitive Learning** | Directly aligns model performance with business objectives; requires no modification to the dataset. | Requires an accurate cost matrix which can be subjective; not natively supported by all algorithms. |
