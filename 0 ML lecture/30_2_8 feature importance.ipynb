# **Feature Importance using Random Forest and Decision Trees**

### **Using Random Forest for feature importance**

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('mnist1.5k.csv')
print(df.shape)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

sns.heatmap(X.iloc[5].values.reshape(28, 28))
```

```python
rf = RandomForestClassifier()
rf.fit(X, y)

print(rf.feature_importances_.shape)
print(rf.feature_importances_)          # will print the importance of all the features in the dataset
```

```python
sns.heatmap(rf.feature_importances_.reshape(28, 28))
```

### **Using Decision Trees for feature importance**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.tree import plot_tree
```

```python
X,y = make_classification(
    n_samples=5, 
    n_classes=2,                       
    n_features=2, 
    n_informative=2, 
    n_redundant=0,
    random_state=0
)

clf = DecisionTreeClassifier()
clf.fit(X,y)

plot_tree(clf)

```

```python
print(clf.feature_importances_)
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)
```

```python
X,y = make_classification(
    n_samples=15, 
    n_classes=2,                       
    n_features=2, 
    n_informative=2, 
    n_redundant=0,
    random_state=0
)

clf = DecisionTreeClassifier()
clf.fit(X,y)
plot_tree(clf)

```
