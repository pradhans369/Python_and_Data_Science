# **Feature Importance using Random Forest and Decision Trees**

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('mnist1.5k.csv')
print(df.shape)
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

sns.heatmap(X.iloc[5].values.reshape(28, 28))
```

```python
rf = RandomForestClassifier()
rf.fit(X, y)

print(rf.feature_importances_.shape)
print(rf.feature_importances_)          # will print the importance of all the features in the dataset
```

```python
sns.heatmap(rf.feature_importances_.reshape(28, 28))
```
