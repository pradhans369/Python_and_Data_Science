# **Logistic Regression**

In logistc reg we dont use the $y = mx + b$, rather we use the formula $ax + by + c = 0$


2D = $ax + by + c = 0$

3D = $ax + by + cz + d = 0$

And the formula changes as we move towards the higher dimension data.

In the first epoch you have to take all the values of a, b, and c at random.

### 1. Sigmoid (Logistic) Function
The linear output $z = \beta_0 + \beta_1x_1 + \dots$ is passed through the **Sigmoid Function** to map any real value into a probability range between 0 and 1:
$$\sigma(z) = \frac{1}{1 + e^{-z}}$$

### 2. Hypothesis Representation
The prediction is defined as the probability that $y=1$ given $x$:
$$h_\theta(x) = P(y=1 | x; \theta) = \frac{1}{1 + e^{-(\theta^T x)}}$$

### 3. Decision Boundary
To categorize the output into classes, we apply a threshold (typically 0.5):
*   If $h_\theta(x) \geq 0.5 \implies y = 1$
*   If $h_\theta(x) < 0.5 \implies y = 0$

### 4. Cost Function: Log Loss
We use **Binary Cross-Entropy** because it is convex for logistic regression, unlike Mean Squared Error:
$$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right]$$

### 5. Gradient Descent
To minimize the cost function, weights are updated iteratively:
$$\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)}$$
