```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# importing KMeans class from 32_5
from Kmeans_32_5 import KMeans


import warnings
warnings.filterwarnings("ignore")
```

```python

# making a dataset
centroids = [(-5, -5), (5, 5)]
cluster_std = [1,1]

X, y = make_blobs(
    n_samples=1000,
    n_features=2,
    cluster_std=cluster_std,
    centers=centroids,
    random_state=2
)

# plotting the figure
plt.figure(figsize=(18, 4))

sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
plt.show()
```

```python
km = KMeans(n_clusters=4, max_iter=100)
y_means = km.fit_predict(X)

plt.figure(figsize=(18, 4))
sns.scatterplot(x=X[y_means==0,0], y=X[y_means==0,1], color='red')
sns.scatterplot(x=X[y_means==1,0], y=X[y_means==0,1], color='green')


plt.tight_layout()
plt.show()
```
