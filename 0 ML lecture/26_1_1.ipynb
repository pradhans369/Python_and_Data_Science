# Geometric Intuition: Logistic Regression

**Logistic Regression** is a supervised machine learning algorithm used for **classification** (specifically binary classification), though its name suggests regression. It is used to calculate the probability of a data point belonging to one of two categories (usually represented as 0 and 1).

The primary goal of Logistic Regression from a geometric perspective is to find the **best separating hyperplane** (a line in 2D, a plane in 3D, or a hyperplane in higher dimensions) that divides two classes of data.

---

### 1. The Hyperplane (Decision Boundary)
Imagine a 2D space with two classes: **Class 1 (Positive)** and **Class 0 (Negative)**. Logistic Regression attempts to find a linear boundary defined by:

$$w^T x + b = 0$$

*   **$w$ (Weight Vector):** Geometrically, $w$ is a vector **perpendicular (normal)** to the decision boundary.
*   **Direction:** The vector $w$ points toward the region where Class 1 is predicted.

---

### 2. Signed Distance from the Hyperplane
For any specific data point $x_i$, the expression $w^T x_i + b$ represents the **signed distance** from the hyperplane to that point.

*   **Positive Side ($w^T x_i + b > 0$):** The point belongs to Class 1.
*   **Negative Side ($w^T x_i + b < 0$):** The point belongs to Class 0.
*   **Magnitude:** The absolute value $|w^T x_i + b|$ tells us how far the point is from the line. Points further away signify higher confidence in classification.

---

### 3. Maximizing Classification Correctness
The algorithm aims to ensure that:
1.  **Positive points** ($y_i = +1$) have a **positive distance**.
2.  **Negative points** ($y_i = -1$) have a **negative distance**.

We can represent mathematical "correctness" for a single point as the product:
$$Y_i \cdot (w^T x_i + b)$$

*   If this product is **Positive**, the model is correct.
*   If this product is **Negative**, the model is incorrect.

**Geometric Goal:** Rotate and shift the hyperplane to maximize the sum of these distances for all points while penalizing misclassifications.

---

### 4. The Sigmoid "Squashing" Function
In geometry, distances can range from $-\infty$ to $+\infty$. However, probabilities must range from $0$ to $1$. The **Sigmoid Function** acts as a bridge:

$$P(y=1|x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$$

**Geometric Intuition of Sigmoid:**
- **At the boundary ($dist = 0$):** Prob is exactly $0.5$.
- **Moving away into Positive region:** Prob rapidly approaches $1$.
- **Moving away into Negative region:** Prob rapidly approaches $0$.

---

### Summary Table

| Feature | Geometric View | Probabilistic View |
| :--- | :--- | :--- |
| **Equation** | Hyperplane Equation ($w^T x + b = 0$) | Log-Odds Ratio |
| **Point on Boundary** | Distance = 0 | Probability = 0.5 |
| **Confidence** | Large Distance from Hyperplane | Probability closer to 1 or 0 |
| **Weights ($w$)** | Vector normal to boundary | Feature Importance |

> **Key takeaway:** Geometrically, Logistic Regression is a **Linear Classifier**. It assumes the two classes can be separated by a straight line or plane. If the classes are non-linearly separable (like concentric circles), standard Logistic Regression will fail.


# Multi-Class Logistic Regression: Geometric View

Since a single hyperplane can only divide space into two halves, Logistic Regression uses specific strategies to handle 3 or more classes.

### 1. One-vs-Rest (OvR) Strategy
In One-vs-Rest, the model trains **one binary classifier per class**. 

*   **Geometric Intuition:** If you have classes $\{A, B, C\}$, the model draws 3 separate lines:
    1.  Line A vs. {B, C}
    2.  Line B vs. {A, C}
    3.  Line C vs. {A, B}
*   **Decision Rule:** For a new point, the model calculates the probability from all 3 classifiers. The point is assigned to the class that yields the **maximum probability** (the furthest positive distance from its respective line).

### 2. Multinomial (Softmax) Strategy
Instead of multiple independent lines, this approach treats the classes as a single system using the **Softmax Function**.

*   **Geometric Intuition:** This creates a "competition" between class hyperplanes. The boundary between any two classes is still a hyperplane, but they all meet at a common intersection point.
*   **Regions:** This creates distinct "territories" in the feature space. Any point falling into "Region A" is classified as Class A because its distance-based score for A is higher than its scores for B or C.

---

| Feature | Binary Logistic Reg | Multi-Class (OvR) | Multi-Class (Softmax) |
| :--- | :--- | :--- | :--- |
| **Number of Lines** | 1 Line | $N$ Lines (for $N$ classes) | $N$ Score-planes |
| **Squashing Function** | Sigmoid | Sigmoid (applied $N$ times) | Softmax |
| **Space Partition** | Divided into 2 | Overlapping Binary Splits | Divided into $N$ Regions |

> **Key Takeaway:** Whether it uses OvR or Softmax, Logistic Regression remains a **Linear Classifier**. It can only separate multiple classes if they can be divided by a series of straight lines.
