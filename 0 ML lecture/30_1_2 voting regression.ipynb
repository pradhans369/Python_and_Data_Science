# **Voting Regression**

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, VotingRegressor
from sklearn.svm import SVC, SVR
# SVM - support vector machine    (for classification problems)
# SVR - support vector regression (for regression problems)


import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes  # loading the diabetes regression dataset
X, y = load_diabetes(return_X_y=True)


print(X.shape)
print(y.shape)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()

estimators = [('lr', lr), ('dt', dt), ('svr', svr)]

for i in estimators:
    scores = cross_val_score(i[1], X, y, scoring='r2', cv=10)
    print(i[0], np.round(np.mean(scores), 2))
```

```python
vr = VotingRegressor(estimators)           # we use voting regressor for regression problems
scores = cross_val_score(vr, X, y, scoring='r2', cv=10)
print("Voting Regressor", np.round(np.mean(scores), 2))
```

```python
for i in range(1,4):
  for j in range(1,4):
    for k in range(1,4):
      vr = VotingRegressor(estimators,weights=[i,j,k])
      scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
      print("For i={},j={},k={}".format(i,j,k),np.round(np.mean(scores),2))
```

### **Using different models of decision tree regressor**

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=3)
dt3 = DecisionTreeRegressor(max_depth=5)
dt4 = DecisionTreeRegressor(max_depth=7)
dt5 = DecisionTreeRegressor(max_depth=None)

estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]

for i in estimators:
  scores = cross_val_score(i[1],X,y,scoring='r2',cv=10)
  print(i[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```
