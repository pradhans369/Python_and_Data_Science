```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_classification(
    n_samples=50,
    n_features=4,
    n_classes=4,
    n_informative=3,
    n_redundant=0,
    random_state=42
)

dt = DecisionTreeClassifier()
dt.fit(X, y)

print(dt.feature_importances_)

import matplotlib.pyplot as plt

plt.figure(figsize=(18, 8)) 
plot_tree(dt);          # remove the semicolon here, and you will see the changes
```

```python
rf = RandomForestClassifier()
rf.fit(X, y)

print(rf.feature_importances_)                      # will print the best features out of all the decision trees
print(rf.estimators_)                               # will print all the decision trees
print(rf.estimators_[0].feature_importances_)       # prints an array of numbers that tells you how important each column (feature) was to a specific decision tree in your Random Forest.
print(rf.estimators_[1].feature_importances_)




```

```python
X, y = make_classification(
    n_samples=50,
    n_features=4,
    n_classes=4,
    n_informative=3,
    n_redundant=0,
    random_state=42
)

dt = RandomForestClassifier(
    n_estimators=2,
    random_state=42
)
dt.fit(X, y)

print(dt.feature_importances_)
```
