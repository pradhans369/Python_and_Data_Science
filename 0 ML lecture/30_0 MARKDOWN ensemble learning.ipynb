# **Ensemble Learning**

Ensemble -- means a group of something

### Ensemble Learning
**Definition**: Ensemble learning is a machine learning technique that combines the predictions of multiple individual ML models (often called "base learners" or "weak learners") to produce a single, more accurate, and robust predictive model.

**Explanation**: 
The core idea is the "wisdom of the crowd"—the collective decision of a group of models is generally more reliable than the decision of any single individual model. By aggregating multiple models, ensemble methods help to:
*   **Reduce Variance (Bagging)**: Decreases overfitting by averaging results from models trained on different data subsets (e.g., Random Forest).
*   **Reduce Bias (Boosting)**: Improves accuracy by training models sequentially, where each new model focuses on correcting the errors of its predecessor (e.g., XGBoost).
*   **Improve Generalization (Stacking)**: Combines different types of models using a meta-learner to find the most effective way to weight their predictions.

### **Stacking**

Base models are different (using different ML algorithms) on same dataset. The only difference is you add one more model, that use predictions of all the previous models as input features to make final prediction.

Level 0 models are provided with the same dataset.

Stacking (Stacked Generalization) involves training a meta-model to combine the predictions of several base learners.

1.  **Level-0 Models (Base Models):** Multiple different machine learning models are trained on the original dataset.
2.  **Level-1 Model (Meta-Model):** A final model that uses the predictions of the Level-0 models as input features to make the final prediction.

Here the level 1 model gives more weight to the model with highest accuracy, and less weight to the model with lower accuracy.


**Models used in level 0 and level 1**

*   **Base Models (Level-0):** To maximize the benefit of stacking, use a diverse set of algorithms that capture different patterns in the data:
    *   **Random Forest / Extra Trees** (Bagging)
    *   **Support Vector Machines (SVM)** (Kernel-based)
    *   **K-Nearest Neighbors (KNN)** (Instance-based)
    *   **XGBoost / LightGBM / CatBoost** (Boosting)
*   **Meta-Model (Level-1):** A simple model is preferred to aggregate predictions without introducing additional complexity or overfitting:
    *   **Logistic Regression** (for classification tasks)
    *   **Linear Regression** (for regression tasks)
    *   **Ridge / Lasso Regression** (to provide regularization)



### **Bagging**

It stands for **Bootstrap Aggregation**. Here you have same models at level 0, but you provide them with different data.

As they are provided with different data, they will be predicting different accuracies. A eg of it is Random Forest.


**Models Used in Bagging**

While **Decision Trees** are the most common base learners used in bagging (forming a **Random Forest**), the technique is model-agnostic and can be applied to various algorithms:

1.  **Bagged Decision Trees:** The standard approach where deep, unpruned trees are used to reduce variance.
2.  **Bagged KNN:** Useful for stabilizing the predictions of K-Nearest Neighbors by training on different subsets.
3.  **Bagged SVMs:** Improves the robustness of Support Vector Machines, especially on noisy datasets.
4.  **Bagged Neural Networks:** Reduces the sensitivity of neural networks to initial weights and specific training instances.

In general, bagging is most effective when applied to **high-variance, low-bias models** (complex models that are prone to overfitting).




### **Boosting**

Its a **most powerful and highly accurate** technique compared to the previous techniques.

Here you are also provided with the same models at level 0. 

Boosting is a sequential ensemble technique where models are trained one after another. Each subsequent model attempts to correct the errors made by the previous models in the sequence.


**Key Characteristics:**
- **Sequential Learning:** Unlike Bagging, models are built dependency-wise.
- **Error Correction:** It assigns higher weights to observations that were incorrectly predicted by previous models.
- **Bias Reduction:** Primarily focuses on reducing bias and improving the predictive power of weak learners.

**Popular Boosting Algorithms:**
1. **AdaBoost (Adaptive Boosting):** Increases the weights of misclassified data points so the next model focuses more on them.
2. **Gradient Boosting (GBM):** Trains new models to predict the residual errors (the difference between actual and predicted values) of the previous models.
3. **XGBoost (Extreme Gradient Boosting):** An optimized implementation of gradient boosting designed for high efficiency and performance.
4. **LightGBM / CatBoost:** Modern variations optimized for large datasets and categorical feature handling.

### Advantages of Ensemble Learning
1. **Improved Accuracy**: Combines multiple models to achieve better predictive performance than any single constituent model.
2. **Robustness**: Reduces the risk of making a poor choice by averaging out individual model errors.
3. **Reduced Overfitting**: Techniques like Bagging (e.g., Random Forest) help decrease the variance of the model.
4. **Reduced Bias**: Techniques like Boosting help improve the performance of weak learners by focusing on previously misclassified instances. It maintains low bias and low variance.

### Disadvantages of Ensemble Learning
1. **Increased Computational Cost**: Training and maintaining multiple models requires more time, memory, and processing power.
2. **Reduced Interpretability**: Ensemble models are often "black boxes," making it difficult to explain why a specific prediction was made compared to a single decision tree or linear regression.
3. **Inference Latency**: Predicting new instances takes longer because the input must pass through all models in the ensemble.
4. **Complexity in Deployment**: Managing the lifecycle of multiple models in a production environment is more challenging.


### When to Use Ensemble Learning

**To be frank, there is no reason to use ensemble learning. You should always use ensemble learning. Despite having high computational cost.** 

**Use it when you are done with all the data cleaning, outlier handling and feature selection thing.**


1. **High Accuracy is Critical**: When the primary goal is to achieve the highest possible predictive performance, such as in machine learning competitions or high-stakes applications.
2. **Model Overfitting (High Variance)**: Use Bagging techniques (like Random Forest) when a single model is too complex and captures noise in the training data.
3. **Model Underfitting (High Bias)**: Use Boosting techniques (like Gradient Boosting or AdaBoost) when a single model is too simple to capture the underlying patterns in the data.
4. **Complex Data Patterns**: When the relationship between features and the target variable is highly non-linear and cannot be easily captured by a single traditional model.
5. **Sufficient Computational Resources**: When the project budget and infrastructure can handle the increased training time and memory requirements.
