# **MSE**

## MSE (Mean Squared Error) in Regression

Mean Squared Error is a standard metric used to measure the performance of a regression model by calculating the average of the squared differences between predicted and actual values.


### 1. The Core Formula
$$MSE = \frac{1}{n} \sum (Actual - Predicted)^2$$

It calculates the average of the **squared** differences between what your model predicted and what really happened.


### 2. Why do we Square the error? ($^2$)
Squaring is not just arbitrary; it serves two vital purposes:

* **Eliminates Negative Signs:** If you predict $+2$ too high and another time $-2$ too low, simple addition makes the total error $0$. Squaring makes them $4$ and $4$, so they add up to $8$, revealing the true error.
* **Punishes Big Mistakes:** An error of $2$ becomes $4$ when squared, while an error of $10$ becomes $100$. A mistake that is $5x$ bigger gets penalized **$25x$ harder**, forcing the model to focus on fixing large errors.


### 3. Pros & Cons

| Pros | Cons |
| :--- | :--- |
| **Math Friendly:** The curve ($x^2$) is smooth and differentiable, making it perfect for Gradient Descent to find the minimum efficiently. | **Outlier Sensitive:** One massive outlier creates a huge squared error, potentially dragging the entire model off-course to fix just that one point. |
| **Unique Solution:** For standard linear regression, MSE guarantees a single optimal solution (global minimum). | **Unit Mismatch:** The output unit is "Squared" (e.g., Dollars²), which is unintuitive to read until you take the square root (RMSE). |


### 4. The Outlier Effect
Due to the squaring, a single outlier (e.g., a prediction off by $1,000$) generates an error of **$1,000,000$**. The MSE algorithm sees this massive number and "panics," shifting the regression line drastically to reduce it. This often sacrifices accuracy on the $99\%$ of normal data points just to accommodate the $1\%$ (the outlier).
