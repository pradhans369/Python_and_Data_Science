## **Entropy for classification variables**

**Decision Tree Entropy**


Entropy is nothing but the measure of disorder. Or you can also call it the measure of purity/impurity.

In the context of decision trees, Entropy is a mathematical measure used to calculate the degree of disorder, impurity, or randomness in a dataset.

When building a decision tree, entropy helps the algorithm decide how to split the data. The goal is to move from a state of high entropy (messy, mixed data) to low entropy (clean, organized groups).


Entropy is calculated using the probabilities of the different classes present in the data:

$$H(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

- $H(S)$: The entropy of the set $S$.'
- $p_i$: The proportion (probability) of examples belonging to class $i$.
- $\log_2$: The logarithm to the base 2 is used because entropy is measured in "bits".

**Observation**
- more the uncertanity more is entropy
- for a 2 class problem the min entropy is 0 and the max is 1
- for more than 2 classes the min entropy is 0 but the max can be greater than 1
- both log_2 or log_e can be used to calculate entropy
