```python
import pandas as pd
from sklearn.datasets import load_iris      # iris is a built in data set in sklearn    

"""   
iris is not a normal DataFrame — it's a special object called a Bunch (like a dictionary).

iris contains several items :
    iris.data           → the feature values (numeric data)
    iris.feature_names  → column names (like sepal length, petal width, etc.)
    iris.target         → the class labels (0, 1, 2 for three flower species)
    iris.target_names   → the species names (setosa, versicolor, virginica)

"""


# storing the iris data into the "iris", and then storing it in "data" and converting it into a data frame and then storing it again into the "df"
iris = load_iris()
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)       # storing iris.feature_names into the variable "columns"
# iris
df.head()
```

Pre processing

```python
# originally the arrays were in 1D format, here we have converted them into 2D format
# bcoz the linear regression model expects us give it 2D input data

x = df['petal length (cm)'].values.reshape(-1, 1)
y = df['petal width (cm)'].values.reshape(-1, 1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)


"""     
x_train     → inputs used to train the model
y_train     → outputs used to train the model
x_test      → inputs used to test the model
y_test      → actual outputs for comparison

"""



print(x_train)
print('********************************************************')
# print(x_test)
print(y_train)
print('********************************************************')
# print(y_test)
```

Training the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train) 
print("Coefficient (m):", regressor.coef_)
print("Intercept (c):", regressor.intercept_)
```

Prediction

```python
# now the machine has learned the pattern btwn input and output from above codes
# now you can enter any input in the predict function to get any output, you can even enter an array or a list to get all the outputs for all numbers inside the array
# you can even enter a one single number    --      MAKE SURE THAT WHILE ENTERING A NUMBER, YOU MUST WRITE IT IN A 2D FORMAT, OR ELSE WILL THROW ERROR

regressor.predict([[21]])
```

```python
score = regressor.predict(x_test)
print(score[:5])
```

```python
def calculate(slope, intercept, value):
    return slope * value + intercept        # y = mx + c



score2 = calculate(regressor.coef_, regressor.intercept_, 21)
print(f"THE SCORE IS : {score2}")

# the ans you will get will exactly match this code "regressor.predict([[21]])"
# bascially this is another way write that code
```

Printing the test data in a pandas data frame format

```python
# x_train     → inputs used to train the model
# y_train     → outputs used to train the model
# x_test      → inputs used to test the model
# y_test      → actual outputs for comparison

import numpy as np
y_pred = regressor.predict(x_test)
y_pred = np.round(y_pred)

results = pd.DataFrame({'ACTUAL DATA' : y_test.squeeze(), 'PREDICTION ACQUIRED' : y_pred.squeeze()})        # printing the data in a data frame format
results
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"MAE : {mae}")            # It calculates the average of the absolute differences between the actual values and the predicted values. [“On average, how much did my model’s prediction miss the real value by?”]
# how much y pred deviates from y test (it tells us the errors, the lower the better), the min mae can be 0 and the max in no limit

print(f"MSE : {mse}")            # It squares the differences first, then takes the average. [so if your model makes one big error, MSE increases a lot.]
# finds the difference btwn actual data and prediction data, then squares the each difference, and then takes the average

print(f"RMSE : {rmse}")           # It’s just the square root of MSE, which brings the error back to the same unit as your target variable (like marks, salary, etc.) [“What’s the average size of my error (but in the same units as my data)?”]


# MAE tells the average size of errors,
# MSE tells the average of squared errors (penalizes big mistakes more),
# and RMSE is the square root of MSE, showing the error in the same units as the target.

```
