```python

import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import scipy.cluster.hierarchy as shc               
from sklearn.cluster import AgglomerativeClustering # used for aglomerative clustering
```

```python
df = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
print(df.shape)
df.head(10)
```

```python
data = df.iloc[:,3:5]
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()
```



If you want a truly automatic way of choosing the clusters(without looking at a plot), you would use something like the `Silhouette Score` or the `Elbow Method`, where you run the algorithm multiple times and calculate which one has the best math results.

The dendrogram shows you the hierarchy of how data points merged. It's up to you to "cut" the tree at a height that makes sense for your data!

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    metric='euclidean',
    linkage='ward'
)

labels_ = cluster.fit_predict(data)

labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:, 0], data.iloc[:,1], c=cluster.labels_, cmap='rainbow')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
