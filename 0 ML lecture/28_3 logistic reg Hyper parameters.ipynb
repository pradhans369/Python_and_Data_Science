# Logistic Regression Hyperparameters (Scikit-Learn)

In `sklearn.linear_model.LogisticRegression`, hyperparameters are the "knobs" you turn to control how the model learns, prevents overfitting, and handles complex data.


### 1. `penalty` (The Type of Regularization)
Regularization prevents the model from becoming too complex (overfitting) by penalizing large weights.
*   **`'l2'` (Default) [Ridge]:** Adds the "square" of the weights to the loss function. It shrinks coefficients but rarely makes them exactly zero.
*   **`'l1'` [Lasso]:** Adds the "absolute value" of weights. It can shrink some coefficients to **exactly zero**, effectively performing feature selection.
*   **`'elasticnet'`:** A combination of both L1 and L2. Requires the `l1_ratio` parameter.
*   **`None`:** No penalty is applied.


### 2. `C` (Inverse Regularization Strength)
This is arguably the most important hyperparameter to tune.
*   **Formula:** $C = 1 / \lambda$ (where $\lambda$ is the strength of the penalty).
*   **Small `C` (e.g., 0.01):** High regularization. The model is forced to keep weights small. This helps if you are **overfitting**.
*   **Large `C` (e.g., 100):** Low regularization. The model tries to fit the training data as closely as possible. This helps if you are **underfitting**.


### 3. `solver` (The Optimization Algorithm)
This is the "engine" used to find the best weights. Different solvers work better for different data sizes and penalties.
*   **`'lbfgs'` (Default):** Good for small to medium datasets. Supports only L2 or no penalty.
*   **`'liblinear'`:** Great for small datasets. The only one that supports L1 on small data. Not great for multi-class (use one-vs-rest).
*   **`'saga'`:** The best choice for **large datasets**. Supports all penalties (L1, L2, ElasticNet).
*   **`'newton-cg'`:** Computationally expensive but very accurate for multiclass problems.


### 4. `multi_class` (Handling Multiple Labels)
*   **`'ovr'` (One-vs-Rest):** Fits a binary model for each class.
*   **`'multinomial'` (Softmax):** Fits a single model that predicts probabilities across all classes (use this if the classes are mutually exclusive).


### 5. `max_iter` (Maximum Iterations)
The solver is an iterative process (Gradient Descent).
*   **Purpose:** How many steps the solver takes to find the minimum point of the loss function.
*   **Default:** 100.
*   **Common Issue:** If you see a warning saying "ConvergenceWarning: lbfgs failed to converge," you should increase this to `1000` or higher.


### 6. `class_weight` (Handling Imbalanced Data)
*   **`None`:** All classes are treated as equally important.
*   **`'balanced'`:** Automatically adjusts weights inversely proportional to class frequencies. Very helpful if you have 90% "No" and 10% "Yes" in your data.

### Summary Table for Selection
| Parameter | Best for Overfitting | Best for Large Data | Best for Feature Selection |
| :--- | :--- | :--- | :--- |
| **Penalty** | `l2` or `l1` | `l2` | `l1` |
| **C** | Small value (0.01) | Large value (10) | N/A |
| **Solver** | Any | `saga` | `liblinear` or `saga` |
| **Max_iter** | N/A | Increase to >1000 | N/A |


### Example Implementation
```python
from sklearn.linear_model import LogisticRegression
# Robust setup for a typical project
model = LogisticRegression(
    penalty='l2',        # Type of penalty
    C=1.0,               # Default strength
    solver='lbfgs',      # Standard engine
    max_iter=1000,       # Increased to avoid convergence errors
    class_weight='balanced', # Handles imbalanced classes
    n_jobs=-1            # Use all CPU cores
)
