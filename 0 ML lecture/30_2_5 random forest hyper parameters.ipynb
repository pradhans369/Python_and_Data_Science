# **Random Forest Hyper Parameters**


### Random Forest Hyperparameters for Classification

| Hyperparameter | Description | Default | Typical Range/Values |
| :--- | :--- | :--- | :--- |
| **`n_estimators`** | Number of trees in the forest. More trees usually improve performance but increase computation time. | `100` | `100 - 1000` |
| **`criterion`** | Function to measure the quality of a split. | `"gini"` | `"gini"`, `"entropy"`, `"log_loss"` |
| **`max_depth`** | Maximum depth of each tree. Helps prevent overfitting by limiting tree complexity. | `None` | `5 - 30` or `None` |
| **`min_samples_split`** | Minimum number of samples required to split an internal node. | `2` | `2 - 20` |
| **`min_samples_leaf`** | Minimum number of samples required to be at a leaf node. Smoothes the model. | `1` | `1 - 10` |
| **`max_features`** | Number of features to consider when looking for the best split. | `"sqrt"` | `"sqrt"`, `"log2"`, `None`, or float |
| **`bootstrap`** | Whether bootstrap samples are used when building trees (sampling with replacement). | `True` | `True`, `False` |
| **`oob_score`** | Whether to use out-of-bag samples to estimate the generalization score. | `False` | `True`, `False` |
| **`n_jobs`** | Number of CPU cores to use for parallel processing. `-1` uses all processors. | `None` | `-1` |
| **`random_state`** | Controls the randomness of the bootstrapping and feature selection for reproducibility. | `None` | Integer (e.g., `42`) |
| **`class_weight`** | Weights associated with classes (useful for imbalanced datasets). | `None` | `"balanced"`, `"balanced_subsample"`, or dict |



### Random Forest Hyperparameters for Regression

| Hyperparameter | Description | Default | Typical Range/Values |
| :--- | :--- | :--- | :--- |
| **`n_estimators`** | Number of trees in the forest. More trees usually improve performance but increase computation time. | `100` | `100 - 1000` |
| **`criterion`** | Function to measure the quality of a split. | `"mse"` | `"mse"`, `"mae"`, `"friedman_mse"` |
| **`max_depth`** | Maximum depth of each tree. Helps prevent overfitting by limiting tree complexity. | `None` | `5 - 30` or `None` |
| **`min_samples_split`** | Minimum number of samples required to split an internal node. | `2` | `2 - 20` |
| **`min_samples_leaf`** | Minimum number of samples required to be at a leaf node. Smoothes the model. | `1` | `1 - 10` |
| **`max_features`** | Number of features to consider when looking for the best split. | `"sqrt"` | `"sqrt"`, `"log2"`, `None`, or float |
| **`bootstrap`** | Whether bootstrap samples are used when building trees (sampling with replacement). | `True` | `True`, `False` |
| **`oob_score`** | Whether to use out-of-bag samples to estimate the generalization score. | `False` | `True`, `False` |
| **`n_jobs`** | Number of CPU cores to use for parallel processing. `-1` uses all processors. | `None` | `-1` |
| **`random_state`** | Controls the randomness of the bootstrapping and feature selection for reproducibility. | `None` | Integer (e.g., `42`) |
| **`max_samples`** | Number of samples to draw from X to train each base estimator. | `None` | `None`, `int`, or float |
