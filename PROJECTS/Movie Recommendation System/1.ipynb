### Formatting of features

```python
print(df.iloc[0]['GENRES'])
print(df.iloc[0]['KEYWORDS'])
print(df.iloc[0]['CAST'])
print(df.iloc[0]['CREW'])
```

As we can above the above four features are in a list of multiple dictionaries format, hene we have to convert them into a proper format.

```python
type(df.iloc[0]['GENRES'])
```

As you can clearly see, the 'GENRES' is in list format but its type not in string format, So first we have to convert it into a list format.

```python
import ast

# for GENRE, KEYWORD
def converter(obj):
    temp = []

    for i in ast.literal_eval(obj):
        temp.append(i['name'])
    return temp

# for CAST
def converter2(obj):
    temp=[]
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            temp.append(i['name'])
            count += 1
        else:
            break
    return temp

# for CREW
def converter3(obj):
    temp = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            temp.append(i['name'])
            break
    return temp
```

- **GENRES  :** tell about from which genre the movie belongs
- **KEYWORDS :** keywords used to search the movie
- **CAST :** actors hired for the movie
    - For the cast we will only take the first 3 actors name, bcoz in most of the movie top actors names are mentioned first and rest of the cast is not so much important
- **CREW :** crew hired for the movie
    - We are only taking only where there is a director name used, coz directore name also influences the movie 

```python
df['GENRES'] = df['GENRES'].apply(converter)
df['KEYWORDS'] = df['KEYWORDS'].apply(converter)
df['CAST'] = df['CAST'].apply(converter2)
df['CREW'] = df['CREW'].apply(converter3)

df['CREW']
```

### Using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

# by default vectorizer returns a sparse output, hence we will be converting it into an array
print(cv.fit_transform(df['TAGS']).toarray().shape)
vector = cv.fit_transform(df['TAGS']).toarray()
vector
```

**Observation :** By this we have converted every movie into a vector form.

### Removing the Similar Words

To do so we will be applying **Stemming**.

For eg, multiple words like loved, loving, loves will be replaed with only one word like love.

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(txt):
    y = []
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(ps.stem('dancing'))
print(ps.stem('loving'))
```

```python
df['TAGS'] = df['TAGS'].apply(stem)
df['TAGS']
```

### Choosing the Distance to be used for vector movies

```python
from sklearn.metrics.pairwise import cosine_similarity
print(len(cosine_similarity(vector).shape))
cosine_similarity(vector)
```

**Observation :** Here we have chose Cosine similarity distance instead of Euclidean distanace bcoz, euclidean distance fails in high dimensional data.

- Higher the distance higher will be the dis-similarity btwn the movies.

- Euclidean distance takes the distance btwn two points, but Euclidean distance takes the degree as their distance. Higher the degree higher the dis-similarity btwn the movies.

```python
# storing the similarity in an object
similarity = cosine_similarity(vector)
print(similarity[0].tolist() )          # printing the similarity of the first movie with the rest of the movies
print(sorted(similarity[0].tolist(), reverse=True))
```

**Observation :** As you can clearly see in the list above, the max similarity is 1. It is bcoz a movie's similarity with itself will be the highest.

### Making a function where with one input movie will return five movies with max similarity

```python
def recommend(movie):
    movie_index = df[df['TITLE'] == movie].index[0]             # finds the index of the movie
    distances = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x: x[1])
    
    for i in distances[1:6]:
        print(df.iloc[i[0]]['TITLE'])       # 'df' is the dataframe variable and 'TITLE' is the column name
```
